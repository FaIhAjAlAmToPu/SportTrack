```python
data_yaml_content = """
train: /kaggle/working/Football-Detection-2/train/images
val: /kaggle/working/Football-Detection-2/valid/images
test: /kaggle/working/Football-Detection-2/test/images

nc: 2
names: ['Ball', 'Player']
"""

with open("/kaggle/working/sports_data.yaml", "w") as f:
    f.write(data_yaml_content)
```

```python
from ultralytics import YOLO
model = YOLO("/kaggle/input/sportsdet/pytorch/default/2/footballANDplayer.pt")
```

```python
results = model.val(split="test")
```

```python
metrics = results
print(metrics.box.p, end=", ")  # precision
print(metrics.box.r, end=", ")  # recall
print(metrics.box.f1, end=", ")  # f1
print(metrics.box.map)
print(metrics.box.all_ap, end=", ")
```
