```python
data_yaml_content = """
train: /kaggle/input/soccernet/train
val: /kaggle/input/soccernet/train
test: /kaggle/input/soccernet/test

nc: 2
names: ['Ball', 'Player']
"""

with open("/kaggle/working/sports_data.yaml", "w") as f:
    f.write(data_yaml_content)
```

```python
from ultralytics import YOLO
model = YOLO("/kaggle/input/sportsdet/pytorch/default/1/player_detector_football.pt")
```

```python
results = model.val(split="test")
```

```python
metrics = results
print(metrics.box.p, end=", ")  # precision
print(metrics.box.r, end=", ")  # recall
print(metrics.box.f1, end=", ")  # f1
print(metrics.box.map)
print(metrics.box.all_ap, end=", ")
```
